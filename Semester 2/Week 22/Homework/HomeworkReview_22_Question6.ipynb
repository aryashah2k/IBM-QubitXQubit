```python
import numpy as np
# Importing standard Qiskit libraries
from qiskit import QuantumCircuit, transpile, Aer, IBMQ
from qiskit.tools.jupyter import *
from qiskit.visualization import *
from ibm_quantum_widgets import *

# Loading your IBM Q account(s)
provider = IBMQ.load_account()
from qiskit import QuantumCircuit, execute
from qiskit import IBMQ, Aer
from qiskit.visualization import plot_histogram
import numpy as np
provider = IBMQ.load_account()
```

```python
from qiskit.providers.aer.noise import NoiseModel
import matplotlib.pyplot as plt
```

```python
backend_belem = provider.get_backend('ibmq_belem')
noise_model_belem = NoiseModel.from_backend(backend_belem)
backend_yorktown = provider.get_backend('ibmq_5_yorktown')
noise_model_yorktown = NoiseModel.from_backend(backend_yorktown)
```

```python
qc = QuantumCircuit(5,5)
qc.draw('mpl')
```

```python
qc.cnot(2,4)
```

```python
qc.draw('mpl')
```

```python
qc.measure_all()
```

```python

backend = Aer.get_backend('qasm_simulator')

job_belem = execute(qc, backend, noise_model=noise_model_belem, shots = 400)
results_belem = job_belem.result()
counts_belem = results_belem.get_counts()
#count_000_belem = results.get_counts(qc)
print(counts_belem)
plot_histogram(counts_belem)
    
```

```python
backend = Aer.get_backend('qasm_simulator')

job_yorktown = execute(qc, backend, noise_model=noise_model_yorktown, shots = 400)
results_yorktown = job_yorktown .result()
counts_yorktown = results_yorktown.get_counts()
#count_000_belem = results.get_counts(qc)
print(counts_yorktown)
plot_histogram(counts_yorktown)
```
